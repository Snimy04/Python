## Clean the data

```python
import numpy as np
```

```python
# dealing with null values
# first resort must be to replace null with mean for numeric data  and with mode for categorical data
#deleting all rows or columns which has null values( this must be the last resort)--> df.dropna(axis=0,inplace=True) drops rows with null values
#                                                                                     df.dropna(axis=1,inplace=True) drops columns with null values
#inplace is used to reflect the changes made after droppimg rows/cols with null values, in the original data
#if we dont use inplace=True the rows/cols with null values will be deleted temoporarily but no changes will be made to original data
```

```python
dnull = [
            [1,2,3,'M',7],
            [np.nan,5,6,np.nan,7],
            [8,np.nan,9,'F',10],
            [np.nan,np.nan, np.nan,'M',11]
        ]
```

```python
df = pd.DataFrame(dnull, columns=['a','b','c','d','e'])
df
```

## Handle Nulls

#### delete records which have null values

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df = pd.DataFrame(dnull, columns=['a','b','c','d','e'])
df
```

```python
df.dropna(axis=1, inplace=True)
```

```python
df
```

#### Replace nulls with mean or mode

```python
df = pd.DataFrame(dnull, columns=['a','b','c','d','e'])
df
```

```python
avg_a = df['a'].mean()
avg_b = df['b'].mean()
avg_c = df['c'].mean()
```

```python
df['d'].mode()
```

```python
df['d'].mode()[0]
```

```python
mode_d = df['d'].mode()[0]
mode_d
```

```python
df.fillna(0)
```

```python
df.fillna(
    {
        'a':avg_a,
        'b':avg_b,
        'c':avg_c,
        'd':mode_d
    
    }, inplace=True
)
```

```python
df
```

#### merge()

```python
d1 = {'movieid':[1,2,3,4,5],
     'title':['3 Idiots','2 States','Leo','RRR','Pushpa']}
```

```python
movies = pd.DataFrame(d1)
```

```python
movies
```

```python
d2 = {
    'memid':[1,2,3,4],
    'name':['Tom','Tim','Jon','Jim'],
    'movieid':[1,np.nan,3,5]
}
```

```python
members = pd.DataFrame(d2)
members
```

```python
pd.merge(movies,members, on='movieid', how='inner')
```

```python
pd.merge(movies,members, on='movieid', how='left')
```

```python
pd.merge(movies,members, on='movieid', how='right')
```

```python
pd.merge(movies,members, on='movieid', how='outer')
```

#### groupby()

```python
sal = pd.read_csv('Salaries.csv')
```

```python
sal.head()
```

```python
sal['rank'].unique()
```

```python
sal['rank'].value_counts()
```

```python
grp = sal.groupby('rank')
```

```python
grp
```

```python
for x,y in grp:
    print(x)
    print('_' * 10)
```

```python
for x,y in grp:
    print(y)
    print('_' * 55)
```

```python
grp.get_group('AssocProf')
```

```python
#num_cols = sal._get_numeric_data()
num_cols = sal[['rank','phd','service','salary']]
#num_cols
```

```python
ngrp = num_cols.groupby('rank')
```

```python
ngrp.agg('mean').round(2)
```
