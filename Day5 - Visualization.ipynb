# Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

#### a. Line Plot

```python
x = [1,2,3,4,5,6,7]
y = [1,4,9,16,25,36,49]
```

```python
plt.figure(figsize=(4,4))
plt.title('Line Chart', fontsize=14,color='g')
plt.xlabel('X - Values',color='b')
plt.ylabel('Y - Values', color='m')
plt.plot(x,y,color='midnightblue',linestyle='dotted',
        marker = '^', markersize=12, markerfacecolor='y')
plt.show()
```

### scatter plot

### c .barchart

```python
courses=['data science','data anlytics','tableau','power bi','sql']
students=[120,170,80,90,45]
```

```python
plt.figure(figsize=(8,4))
plt.title('students vs courses', fontsize=14,color='g')
plt.xlabel('courses',color='b')
plt.ylabel('students', color='m')
plt.xticks(fontsize=13,rotation=45
          )
plt.bar(courses,students,color='#9C640C',alpha=0.7)
plt.show()
#we can also use html colour codes for fonts as well as bars...type html colour codes in the browser
```

```python
#horizotal bars
plt.figure(figsize=(8,4))
plt.title('students vs courses', fontsize=14,color='g')
plt.ylabel('courses',color='b')
plt.xlabel('students', color='m')
plt.xticks(fontsize=13,rotation=45
          )
plt.barh(courses,students,color='#34495E',alpha=0.7)
plt.show()
```

### d. pie chart

```python
courses=['data science','data anlytics','tableau','power bi','sql']
students=[120,170,80,90,45]
```

```python
plt.pie(students,labels=courses,
        autopct='%0.2f%%',#0.2f% denotes that u want the percentage in 0.2 formatting...this gives two digits after decimal
       textprops={'fontsize':10,'color':'black' },
        explode=[0,0.3,0,0,0.4],shadow=True,pctdistance=0.4,
        labeldistance=0.6,rotatelabels=True) # the order for explode goes as per the order of the data 
plt.show()
```

## catplot

### a. bar

```python
sns.catplot (data=cars,x='Brand',y='Cars Listings',kind='bar', aspect=4/3,palette='muted')
plt.show()
```

```python
sns.catplot (data=cars,y='Brand',x='Cars Listings',kind='bar', aspect=4/3,palette='muted')
plt.show()
```

```python
sns.catplot (data=tips,x='day',y='total_bill',kind='strip', aspect=4/3,palette='muted')
plt.show()
```

```python
sns.catplot (data=tips,x='day',y='total_bill',kind='swarm', aspect=4/3,palette='muted',hue='sex')
plt.show()
'''The only difference between the strip-plot and the swarm-plot is that the swarm-plot spreads out the data points of the 
variable automatically to avoid overlap and hence provides a better visual overview of the data'''
```

```python
sns.catplot (data=tips,x='day',y='total_bill',kind='box', aspect=4/3,palette='muted',hue='sex')
plt.show()
#hue = for each value in sex column give a different colour
```

```python
sns.catplot (data=tips,x='day',y='total_bill',kind='box', aspect=4/3,palette='muted',hue='time')
plt.show()
```
